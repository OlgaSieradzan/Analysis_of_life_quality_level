# Analiza poziomu życia w wojewódzctwie Świętokrzyskim przed i po pandemii

W badaniu pod uwagę brane jest 11 zmiennych diagnostycznych, podzielonych na siedem obszarów poziomu życia:

* Warunki materialne
* Aktywność ekonomiczna i warunki pracy

Wszytskie z tych obszarów zostały uznane za znaczące w związku z poziomem życia w danym powiecie. Poniżej zaprezentowane zostaną statystyki opisowe oraz wykresy gęstości badanych zmiennych.

### Warunki materialne
W skład tej kategorii wchodzą dwie zmienne diagnostyczne: *przeciętne miesięczne wynagrodzenie brutto w zł* oraz *przecietna powierzchnia użytkowa mieszkania na 1 osobę w m^2*. Obydwie zmienne zostały zakwalifikowane jako stymulanty, ponieważ wraz z ich wyższymi odczytami, poziom życia rośnie. 

```python
def decripctions_table(salary):
    data1 = {
        "Średnia" : [salary['2019'].mean(), salary['2023'].mean()],
        "Odch. stan." : [np.std(salary['2019'], ddof = 1), np.std(salary['2023'], ddof = 1) ],
        "Min": [salary['2019'].min(), salary['2023'].min()],
        "Max": [salary['2019'].max(), salary['2023'].max()],
        "Skośność": [sp.stats.skew(salary['2019']), sp.stats.skew(salary['2023'])],
        "Kurtoza": [sp.stats.kurtosis(salary['2019']), sp.stats.kurtosis(salary['2023'])]

    }
    tabela = pd.DataFrame(data = data1, index=['2019', '2020'])
    return tabela
```

**Wynagrodzenia**

```python
salary_deb = decripctions_table(salary)
salary_deb
```

**Powierzchnia mieszkaniowa**

```python
area_deb = decripctions_table(area)
area_deb
```

### Aktywność ekonomiczna i warunki pracy
Do tego obszaru również zaliczono dwie zmienne diagnostyczne: *stopa bezrobocia rejestrowanego w %* oraz *poszkodowanie w wypadkach przy pracy na 1000 pracujących ogółem*. Obydwie zmienne zakfalifikowano jako destymulanty, czyli wraz ze spadkiem danej cechy, spada poziom życia w danym powiecie.
